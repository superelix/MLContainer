```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import MinMaxScaler
import os 
os.chdir("C:\\Users\Aniruddha\Downloads")
```

```python
data1=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
data2=pd.read_csv("0000000000002429_test_titanic_x_test.csv")
```

```python
def f(s):
    if s=="female":
        return 0
    elif s=="male":
        return 1
def g(s):
    if s=="C":
        return 0
    elif s=="S":
        return 1
    elif s=="Q":
        return 2
    else:
        return 3

def change(data):
    data["Sex"]=data.Sex.apply(f)
    data["Embarked"]=data.Embarked.apply(g)
    del data["Ticket"]
    del data["Name"]
    del data["Cabin"]
    return data

x=change(data1)
y=change(data2)
x.head()
```

```python
x.Age.fillna(x.Age.mean(),inplace=True)
y.Age.fillna(y.Age.mean(),inplace=True)
x["pc"]=x["SibSp"]+x["Parch"]+1
y["pc"]=y["SibSp"]+y["Parch"]+1
m=x.drop("Survived",axis=1)
m.head(10)
```

```python
algo=LogisticRegression(max_iter=300,tol=0.001)
algo.fit(m,x.Survived)
y_predict=algo.predict(y)
np.savetxt("boston1.csv",y_predict,fmt="%f")
algo.score(m,x.Survived)
```
